```python
def calculate_split_point(df, independent, target, unique):
    splitpoints=[]
    
    if(len(df) < 2):
        return splitpoints
    df.index = range(len(df))
    info_gain_max = 0
    index = 0
    
    for i in range(1, len(df)-1):
        if df[independent][i] == df[independent][i-1]:
            continue
        curr_info_gain = info_gain(df, target, i, unique)
        if curr_info_gain > info_gain_max:
            info_gain_max = curr_info_gain
            index = i
 
    if index >= len(df) or info_gain_max < 0.01:
        return splitpoints
    
    splitpoints.append(df[independent][index]+0.5)    
    splitpoints.extend(calculate_split_point(df[:index], independent, target, unique))
    splitpoints.extend(calculate_split_point(df[index+1:], independent, target, unique))
    return splitpoints
    
def info_gain(df, target, split_point, unique):
    if(len(df) < 2):
        return 0
    entropy = calculate_entropy(df, target, unique)
    entropy1 = calculate_entropy(df[:split_point], target, unique)
    entropy2 = calculate_entropy(df[split_point:], target, unique)
    
    infogain = entropy - (split_point/len(df)) * entropy1 - ((len(df)-split_point)/len(df)) * entropy2
    return infogain

def calculate_entropy(df, target, unique):
    entropy = 0.0
    for i in unique:
        if len(df) == 0:
            continue
        ratio = len(df[df[target] == i])/len(df)
        if ratio == 0:
            continue
        entropy += -1 * ratio * math.log(ratio)
    return entropy
    
```
